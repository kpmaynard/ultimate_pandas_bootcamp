```python
def popular_meter(x):
  if x<220: 
    return "Relatively Unknown"
  if (x >= 220) & (x < 600): 
    return "Kind of Popular"
  if (x >= 600) & (x < 2000): 
    return "Popular"
  else:
    return "Super Popular"
  
```

```python
popular_meter(0)
```

```python
# 2. apply this function to df.page_views column
```

```python
df.head()
```

```python
df['popularity']=df.page_views.apply(popular_meter)
```
